## Iteration, Indentation, and Blocks
With lists, as with other data types, we can *iterate* through them, i.e. to go through each element one at a time. To do this in Python, we use the **for** statement:

This code snippet goes through each element of the list called **days_of_the_week** and assigns it to the variable **day**. When the program has gone through every element of the list, it exists the block.

(Almost) every programming language defines blocks of code in some way. C, C++, and R, use curly braces {} to define these blocks.

Python uses a colon (":"), followed by indentation level to define code blocks. Everything at a higher level of indentation is taken to be in the same block. Note that while in many programming languages idententation is only recommended for readability, in Python it is part of the language syntax

In the above example the block was only a single line, but we could have had longer blocks as well:

## Functions

Functions are reusable blocks of code that perform a specific task or computation. They help make programs modular and organized by allowing you to define a section of code once which can be called multiple times throughout the program, reducing repetition and improving readability. Functions can take inputs, called parameters, and can return an output using the return statement. By using functions, we can break down complex problems into smaller, manageable pieces, making the code more efficient and easier to debug. Functions are present in the majority of programming languages, with different syntax for their definition. Defining a function in Python is done using the **def** keyword followed by the function name and parentheses, which may include parameters.

The following cell defines a function called *is_prime*, without executing it

```python
def is_prime(number):
    """
    This function checks if a given number is prime.
    A prime number is a number greater than 1 that is only divisible by 1 and itself.
    """
    # Check if the number is less than 2, which are not prime numbers
    if number < 2:
        return False

    # Check for factors from 2 up to the square root of the number
    for i in range(2, int(math.sqrt(number)) + 1):
        if number % i == 0:
            return False

    # If no factors were found, the number is prime
    return True
```

In the following cells we call the function

```python
test_number = 10
res = is_prime(test_number)

if res:
    print(f"{test_number} is a prime number.")
else:
    print(f"{test_number} is not a prime number.")
```

All prime numbers < 50:

```python
for i in range(50):
  if is_prime(i):
    print(i)
```

### Tuples and Dictionaries

Two data structures that are very useful and common in Python are **tuples** and **dictionaries**.

A **tuple** is a sequence object like a list or a string. It is constructed by grouping a sequence of objects together with commas, either without brackets, or with parentheses:

```python
t = (1,2,'hi',9.0)
t
```

Tuples are like lists, in that you can access the elements using indices:

```python
t[1]
```

However, tuples are *immutable*, you can't append to them or change the elements of them:

```python
t.append(7)
```

```python
t[1]=77
```

Tuples are useful anytime you want to group different pieces of data together in an object, but don't want to create a full-fledged class for them. For example, if want to store the Cartesian coordinates of some objects, tuples are a good way to do this:

```python
('Bob',0.0,21.0)
```

Here is a list of this Cartesian coordinates, i.e. a list of tuples:

```python
positions = [
             ('Bob',0.0,21.0),
             ('Cat',2.5,13.1),
             ('Dog',33.0,1.2)
             ]
```

Tuples can be used when functions return more than one value. Suppose we want to compute the smallest x- and y-coordinates of the above list of objects:

```python
def minxy(objects):
    minx = objects[0][1] # These are set to the first x and y elements
    miny = objects[0][2]
    for obj in objects[1:]:
        name, x, y = obj
        if x < minx:
            minx = x
        if y < miny:
            miny = y
    return minx, miny

x, y = minxy(positions)
print(x,y)
```

Here we did two things with tuples. First, we unpacked an object into a set of named variables using *tuple assignment*:

    >>> name,x,y = obj

We also returned multiple values (minx,miny), which were then assigned to two other variables (x,y), again by tuple assignment. This makes what would have been complicated code in another programming languages rather simple.

Tuple assignment is also a convenient way to swap variables:

```python
x,y = 1,2
y,x = x,y
x,y
```

**Dictionaries** are a type of objects alss called "mappings" or "associative arrays" in other languages. Whereas a list associates an integer index with a set of objects, the index in a dictionary is called the *key*, and the corresponding dictionary entry is the *value*. A dictionary can use (almost) anything as the key. Whereas lists are formed with square brackets [], dictionaries use curly brackets {}:

```python
ages = {"Rick": 46, "Bob": 86, "Fred": 21}
print("Rick's age is ", ages["Rick"])
```

There's also a convenient way to create dictionaries without having to quote the keys.

```python
dict(Rick=46,Bob=86,Fred=21)
```

The **len** function works on both tuples and dictionaries:

```python
len(t)
```

```python
len(ages)
```

It is possible to change values or add new values to the dictionary in the same way:

```python
ages['Bob'] = 87    # The key 'Bob' already exists, so that value is updated
ages['Bill'] = 33   # The key 'Bill' doesn't exist, so that value is created
print(ages)
```

```python
ages.pop('Bill')
```

To loop through a dictionary we can use the *keys* method:

```python
for k in ages.keys():
    print(k,ages[k])
```

Note that dictionaries do not conserve order. This means that you shouldn't write code like this:

```python
x_dict = {'x1': 1, 'x2': 11, 'x3': 21}
x_dict.pop('x2')
x_dict['x2'] = 10
y_dict = {'y1': 2, 'y2': 12, 'y3': 22, 'y4':5}

for xkey, ykey in zip(x_dict.keys(), y_dict.keys()):
    # 'zip' allows one to iterate over multiple lists at the same time
    print("{}: {}, {}: {}".format(xkey, x_dict[xkey], ykey, y_dict[ykey]))
```

The above *may or may not* print the 'x' entries along with their corresponding 'y' entries. If you need a dictionary that preserves order, use an [OrderedDict](https://docs.python.org/3/library/collections.html#collections.OrderedDict).

#### Kurtosis and skeweness

Skeweness is a measure of the asymmetry of a distribution

$$
skeweness = 3 \cdot \frac{mean - median}{std}
$$


skeweness > 0 ==> right skew

skeweness < 0 ==> left skew

Kurtosis is a measure of the tailedness of a distribution. Tailedness is how often outliers occur.

There are various formulas to estimate it from a sample of data. The kurtosis method of scipy.stats with default parameters returns the "Excess kurtosis", that is such that:

k > 0 (leptokurtic distribution) has fatter tails than gaussian

k < 0 (platykurtic distribution) has thinner tails than gaussian



```python
from scipy.stats import kurtosis, skew

# Create a list (of dictionaries) to store the results
results = []

for n in num_vars:
    k = kurtosis(df[n])
    s = skew(df[n])
    results.append({'variable': n, 'kurtosis': k, 'skewness': s})

# Convert the results list to a DataFrame
ks_df = pd.DataFrame(results)

ks_df
```

#### Covariance Matrix

$$
Cov(x_i,x_j) = s_{ij} = \frac{1}{n-1}\sum_{k=1}^n (x_{k,i} - \bar{x_i})(x_{k,j}-\bar{x_j})
$$

Positive covariance: the variables increase together

Negative covariance: the variables decrease together

Cov is not bounded and it is influenced by the scale of variables, difficult to compare cov of different pairs of values

```python
# Calculate the covariance matrix
cov_matrix = df[num_vars].cov()
```

#### Correlation matrix
$$
\rho_{ij} = \frac{s_{ij}}{\sqrt{s_{ii}}\sqrt{s_{jj}}}
$$

Bounded in [-1,+1]

A value of +1 indicates a perfect positive linear relationship.

A value of -1 indicates a perfect negative linear relationship.

A value of 0 indicates no linear relationship.

Easier to interpret and compare than covariance, but restricted to linear relationships
